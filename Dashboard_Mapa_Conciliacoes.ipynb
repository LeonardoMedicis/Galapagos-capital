## ğŸ“ Passo 1: Verificar DiretÃ³rio Atual

## ğŸ” Passo 2: Verificar Estrutura do Dashboard

## ğŸ› ï¸ Passo 3: Verificar Node.js

Vamos verificar se o Node.js estÃ¡ instalado:

```python
def verificar_comando(comando):
    """Verifica se um comando estÃ¡ disponÃ­vel"""
    try:
        result = subprocess.run([comando, "--version"], 
                              capture_output=True, text=True, timeout=10)
        if result.returncode == 0:
            return True, result.stdout.strip()
        else:
            return False, result.stderr.strip()
    except (subprocess.TimeoutExpired, FileNotFoundError):
        return False, "Comando nÃ£o encontrado"

# Verificar Node.js
node_ok, node_version = verificar_comando("node")
if node_ok:
    print(f"âœ… Node.js instalado: {node_version}")
else:
    print(f"âŒ Node.js NÃƒO instalado: {node_version}")
    print("ğŸ’¡ Instale Node.js em: https://nodejs.org/")

# Verificar npm
npm_ok, npm_version = verificar_comando("npm")
if npm_ok:
    print(f"âœ… npm instalado: {npm_version}")
else:
    print(f"âŒ npm NÃƒO instalado: {npm_version}")

# Status geral
if node_ok and npm_ok:
    print("\nğŸ‰ Tudo pronto para executar o dashboard!")
else:
    print("\nâš ï¸ Instale Node.js antes de continuar")
```

## ğŸ“¦ Passo 4: Instalar DependÃªncias

Vamos instalar as dependÃªncias do dashboard:

```python
# Navegar para o diretÃ³rio do dashboard
dashboard_dir = "automacao-conciliacoes/mapa-conciliacao-dashboard"

if Path(dashboard_dir).exists():
    print(f"ğŸ“ Navegando para: {dashboard_dir}")
    os.chdir(dashboard_dir)
    
    print(f"ğŸ“ DiretÃ³rio atual: {os.getcwd()}")
    
    # Verificar se node_modules jÃ¡ existe
    if Path("node_modules").exists():
        print("âœ… DependÃªncias jÃ¡ instaladas!")
    else:
        print("ğŸ“¦ Instalando dependÃªncias...")
        print("â³ Isso pode demorar alguns minutos...")
        
        try:
            # Executar npm install
            result = subprocess.run(["npm", "install"], 
                                  capture_output=True, text=True, timeout=300)
            
            if result.returncode == 0:
                print("âœ… DependÃªncias instaladas com sucesso!")
            else:
                print(f"âŒ Erro na instalaÃ§Ã£o: {result.stderr}")
                
        except subprocess.TimeoutExpired:
            print("â° Timeout na instalaÃ§Ã£o - tente executar manualmente")
        except Exception as e:
            print(f"âŒ Erro: {e}")
            
else:
    print("âŒ DiretÃ³rio do dashboard nÃ£o encontrado!")
```

## ğŸš€ Passo 5: Executar Dashboard

Agora vamos executar o dashboard:

```python
import threading
import time
import webbrowser

def executar_dashboard():
    """Executa o dashboard em background"""
    try:
        print("ğŸš€ Iniciando servidor do dashboard...")
        
        # Executar npm run dev
        process = subprocess.Popen(
            ["npm", "run", "dev", "--", "--host", "0.0.0.0"],
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )
        
        # Aguardar alguns segundos para o servidor iniciar
        time.sleep(5)
        
        # Verificar se o processo ainda estÃ¡ rodando
        if process.poll() is None:
            print("âœ… Servidor iniciado com sucesso!")
            print("ğŸŒ Dashboard disponÃ­vel em:")
            print("   â€¢ http://localhost:5173")
            print("   â€¢ http://127.0.0.1:5173")
            
            # Tentar abrir no navegador
            try:
                webbrowser.open('http://localhost:5173')
                print("ğŸŒ Abrindo no navegador...")
            except:
                print("ğŸ’¡ Abra manualmente: http://localhost:5173")
            
            return process
        else:
            stdout, stderr = process.communicate()
            print(f"âŒ Erro ao iniciar servidor:")
            print(f"STDOUT: {stdout}")
            print(f"STDERR: {stderr}")
            return None
            
    except Exception as e:
        print(f"âŒ Erro: {e}")
        return None

# Verificar se estamos no diretÃ³rio correto
if "mapa-conciliacao-dashboard" in os.getcwd():
    print("ğŸ“ DiretÃ³rio correto confirmado!")
    
    # Executar dashboard
    servidor = executar_dashboard()
    
    if servidor:
        print("\nğŸ‰ Dashboard executando!")
        print("\nğŸ“‹ InstruÃ§Ãµes:")
        print("   1. Acesse: http://localhost:5173")
        print("   2. Para parar: Kernel â†’ Interrupt")
        print("   3. Ou feche este notebook")
        
        # Manter o processo vivo
        try:
            servidor.wait()
        except KeyboardInterrupt:
            print("\nâ¹ï¸ Parando servidor...")
            servidor.terminate()
            
else:
    print("âŒ NÃ£o estamos no diretÃ³rio correto do dashboard")
    print(f"ğŸ“ DiretÃ³rio atual: {os.getcwd()}")
    print("ğŸ’¡ Execute as cÃ©lulas anteriores primeiro")
```

## ğŸ¯ Passo 7: Verificar Status do Dashboard

Vamos verificar se o dashboard estÃ¡ funcionando:

```python
import requests
import socket

def verificar_porta(host, porta):
    """Verifica se uma porta estÃ¡ aberta"""
    try:
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        sock.settimeout(3)
        result = sock.connect_ex((host, porta))
        sock.close()
        return result == 0
    except:
        return False

def verificar_dashboard():
    """Verifica se o dashboard estÃ¡ rodando"""
    portas_teste = [5173, 3000, 8080]
    
    print("ğŸ” Verificando se o dashboard estÃ¡ rodando...")
    
    for porta in portas_teste:
        if verificar_porta('localhost', porta):
            print(f"âœ… Servidor encontrado na porta {porta}")
            
            # Tentar acessar via HTTP
            try:
                response = requests.get(f'http://localhost:{porta}', timeout=5)
                if response.status_code == 200:
                    print(f"ğŸŒ Dashboard acessÃ­vel em: http://localhost:{porta}")
                    return f"http://localhost:{porta}"
            except:
                print(f"âš ï¸ Porta {porta} aberta mas nÃ£o responde HTTP")
        else:
            print(f"âŒ Porta {porta} fechada")
    
    print("\nğŸ’¡ Dashboard nÃ£o estÃ¡ rodando. Execute as cÃ©lulas anteriores.")
    return None

# Verificar dashboard
url_dashboard = verificar_dashboard()

if url_dashboard:
    print(f"\nğŸ‰ DASHBOARD FUNCIONANDO!")
    print(f"ğŸ”— Acesse: {url_dashboard}")
    
    # Tentar abrir automaticamente
    try:
        webbrowser.open(url_dashboard)
        print("ğŸŒ Abrindo automaticamente...")
    except:
        print("ğŸ’¡ Copie e cole o link no navegador")
else:
    print("\nâŒ Dashboard nÃ£o estÃ¡ rodando")
    print("ğŸ’¡ Execute as cÃ©lulas anteriores ou use os comandos manuais")
```
